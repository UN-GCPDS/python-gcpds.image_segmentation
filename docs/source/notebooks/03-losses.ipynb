# Losses

### Generalized Cross-Entropy


$$\frac{1}{H*W}\sum_i^H \sum_j^K \sum_k^C 2\frac{1- (\mathbf{M}_{i,j,k}* \hat{\mathbf{M}}_{i,j,k})^q}{q}$$

```python
import numpy as np 
import matplotlib.pyplot as plt 


def GCE(x,q):
    return 2*(1-x**q)/q


x = np.linspace(0,1,100)
qs = np.linspace(0.1,1,10)
GCE_plots = [GCE(x,q) for q in qs]
```

```python
for GCE_plot in GCE_plots:
    plt.plot(x,GCE_plot)


plt.ylim([0,10])
plt.legend(qs)
```
